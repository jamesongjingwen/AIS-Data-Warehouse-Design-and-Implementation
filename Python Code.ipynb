```python
#SQL Codes
queries = {
    "Summary of the latest reported activities of ships": {
        "query": """
            SELECT 
                s.shipname,
                s.shiptype_name,
                l.country,
                st.status_name,
                COUNT(DISTINCT s.ship_id) AS number_of_ships,
                MAX(d.date) AS last_reported_date,
                AVG(m.speed) AS average_speed,
                COUNT(m.message_id) AS message_count
            FROM 
                message_fact m
            JOIN 
                ship_dim s ON m.ship_id = s.ship_id
            JOIN 
                location_dim l ON m.location_id = l.location_id
            JOIN 
                datetime_dim d ON m.datetime_id = d.time_id
            JOIN 
                status_dim st ON m.status_id = st.status_id
            GROUP BY 
                s.shipname, 
                s.shiptype_name, 
                l.country, 
                st.status_name
            ORDER BY 
                last_reported_date DESC,
                number_of_ships DESC;
        """,
        "columns": ["Ship Name", "Ship Type","Country","Status Name","No. of Ships","Last Reported Date","Average Speed","No. of Messages"]
    },
    "Ship details for location with most ships registered in SG": {
        "query": """
            WITH UniqueShipsAtSGPorts AS (
                SELECT 
                    mf.ship_id,
                    ld.ship_lon,
                    ld.ship_lat,
                ld.locode
                FROM message_fact mf
                INNER JOIN location_dim ld ON mf.location_id = ld.location_id
                WHERE ld.LOCODE LIKE 'SG%'
                GROUP BY mf.ship_id, ld.ship_lon, ld.ship_lat, ld.locode
            )

            SELECT 
                sd.ship_id, sd.mmsi, sd.shipname, sd.shiptype_name, sd.country,
                usg.ship_lat,
                usg.ship_lon,
              usg.locode
            FROM ship_dim sd
            JOIN UniqueShipsAtSGPorts usg ON sd.ship_id = usg.ship_id
            ORDER BY sd.ship_id;
        """,
        "columns": ["Ship ID", "MMSI", "Ship Name", "Ship Type", "Country", "Ship Latitude", "Ship Longitude", "LOCODE"]
    },
    "Detailed Log for Specific Ship": {
        "query": """
            SELECT 
                mf.message_id,
                mf.location_id,
                dt.date,
                dt.time_only as time,
                s.description
            FROM 
                message_fact mf
            JOIN 
                status_dim s ON mf.status_id = s.status_id
            JOIN 
                datetime_dim dt ON mf.datetime_id = dt.time_id
            JOIN 
                ship_dim sd ON mf.ship_id = sd.ship_id
            WHERE 
                sd.shipname = '{ship_name}'
            ORDER BY 
                dt.date, dt.time_only;
        """,
        "columns": ["Message Id", "Location ID", "Date", "Time", "Description"]  # Replace with actual column names
    },
    "Estimation of the fuel consumption of each ship": {
        "query": """
            SELECT 
                s.shiptype_name,
                ROUND(AVG(
                    CASE
                        WHEN LEFT(s.shiptype, 1) IN ('4', '6') THEN 0.8 * s.gt * (m.speed * m.speed)
                        WHEN LEFT(s.shiptype, 1) IN ('7', '8') THEN 1.2 * s.gt * (m.speed * m.speed)
                        ELSE s.gt * (m.speed * m.speed)
                    END)
                ) AS estimated_fuel_consumption
            FROM ship_dim s, message_fact m
            WHERE s.ship_id = m.ship_id
            AND s.shiptype_name IS NOT NULL
            AND s.gt IS NOT NULL
            GROUP BY s.shiptype_name
            HAVING AVG(m.speed) > 0
            ORDER BY estimated_fuel_consumption DESC;
            
        """,
        "columns": ["Ship Type","Estimated Fuel Consumption"]
    }
}
```

```python
def open_html_links():
    map_url = "file:///C:/Users/theja/OneDrive/Desktop/MSBA/MSBA%20Course/BT5110/Project/ProjectShip/test.html"
    graph_url = "file:///C:/Users/theja/OneDrive/Desktop/MSBA/MSBA%20Course/BT5110/Project/ProjectShip/type_and_location.html"
    webbrowser.open(map_url)
    webbrowser.open(graph_url)
#Customise this based on your PgAdmin details
def execute_query(query_key, ship_name=None):
    try:
        conn = psycopg2.connect(
            host='localhost',
            dbname='ProjectShip3',
            user='postgres',
            password='', #Change this to your Postgres password
            port='5432'
        )
        cur = conn.cursor()

        
        query = queries[query_key]['query']
        if ship_name:  
            query = query.format(ship_name=ship_name)
        cur.execute(query)

        
        records = cur.fetchall()

       
        for i in results_tree.get_children():
            results_tree.delete(i)

        
        update_treeview_columns(query_key)

        
        for row in records:
            results_tree.insert('', tk.END, values=row)

    except psycopg2.DatabaseError as e:
        
        for i in results_tree.get_children():
            results_tree.delete(i)
        results_tree.insert('', tk.END, values=(str(e),))
    finally:
        
        if cur is not None:
            cur.close()
        if conn is not None:
            conn.close()

def update_treeview_columns(query_key):
    # Clear the current columns
    results_tree['columns'] = ()
    for col in results_tree['columns']:
        results_tree.heading(col, text="")
        results_tree.column(col, width=0, minwidth=0)

    
    new_columns = queries[query_key]['columns']
    results_tree['columns'] = new_columns
    for col in new_columns:
        results_tree.heading(col, text=col.capitalize())
        results_tree.column(col, width=100, anchor='center')

def on_query_selected(event):
    selected_query = query_var.get()
    
    if selected_query == "Detailed Log for Specific Ship":
        ship_name_label.pack(padx=10, pady=5)
        ship_name_dropdown.pack(padx=10, pady=5)
    else:
        ship_name_label.pack_forget()
        ship_name_dropdown.pack_forget()

def on_execute_clicked():
    selected_query = query_var.get()
    selected_ship_name = ship_name_var.get() if ship_name_dropdown.winfo_ismapped() else None
    execute_query(selected_query, selected_ship_name)

    
    if selected_query == "Ship details for location with most ships registered in SG":
        open_html_links()

# Create the main window
root = tk.Tk()
root.title("Maritime Database Query")

# Create a dropdown box for selecting a query
query_label = ttk.Label(root, text="Select a Query:")
query_label.pack(padx=10, pady=5)

query_var = tk.StringVar(root)
query_dropdown = ttk.Combobox(root, textvariable=query_var, width=50, state="readonly")
query_dropdown['values'] = list(queries.keys())
query_dropdown.bind('<<ComboboxSelected>>', on_query_selected)
query_dropdown.pack(padx=10, pady=5)

# Create a label and dropdown for selecting the ship name, initially hidden
ship_name_label = ttk.Label(root, text="Select Ship Name:")
ship_name_var = tk.StringVar(root)
ship_name_dropdown = ttk.Combobox(root, textvariable=ship_name_var, width=50, state="readonly")
ship_name_dropdown['values'] = list_of_ship_names

# Create a frame for the Treeview and scrollbar
tree_frame = ttk.Frame(root)
tree_frame.pack(padx=10, pady=10, fill='both', expand=True)

# Create a Treeview widget for displaying the results
output_label = ttk.Label(root, text="Query Results:")
output_label.pack(padx=10, pady=5)

# Initialize the Treeview with no columns
results_tree = ttk.Treeview(tree_frame, columns=(), show='headings')

# Pack the Treeview
results_tree.pack(side='left', fill='both', expand=True)

# Create a button to execute the query
execute_button = ttk.Button(root, text="Execute Query", command=on_execute_clicked)
execute_button.pack(padx=10, pady=10)

# Start the Tkinter event loop
root.mainloop()
```
